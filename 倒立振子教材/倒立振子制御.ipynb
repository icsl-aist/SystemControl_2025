## 倒立振子台車の運動方程式の導出（ラグランジュ法）

ここでは最初から **角度 $\theta$ を「鉛直上から時計回りを正」** として式を導きます。

---

### 1. 座標と記号の定義

- 水平方向：$x$ 軸（右向き正）  
- 鉛直方向：$y$ 軸（上向き正）  
- カートの位置：$x(t)$  
- カートの質量：$M$  
- 振子の質量：$m$  
- 振子重心までの長さ：$l$  
- カートに加える水平力（右向き正）：$u(t)$  
- 振子の角度：$\theta(t)$  
  - $\theta = 0$：振子が鉛直上向き  
  - 時計回りに倒れると $\theta > 0$  

---

### 2. 振子重心の位置・速度

回転軸はカート上に固定されているとします。  
振子重心の座標 $(X,Y)$ は

$$
X = x - l\sin\theta,\quad
Y = l\cos\theta
$$

時間微分すると

$$
\dot X = \dot x - l\cos\theta\,\dot\theta,\quad
\dot Y = -l\sin\theta\,\dot\theta
$$

したがって速度の 2 乗は

$$
\dot X^2 + \dot Y^2
= \dot x^2 - 2l\cos\theta\,\dot x\dot\theta + l^2\dot\theta^2
$$

---

### 3. 運動エネルギー $T$

カートの運動エネルギー

$$
T_{\text{cart}} = \frac{1}{2}M\dot x^2
$$

振子の運動エネルギー

$$
T_{\text{pend}}
= \frac{1}{2}m(\dot X^2 + \dot Y^2)
= \frac{1}{2}m\bigl(\dot x^2 - 2l\cos\theta\,\dot x\dot\theta + l^2\dot\theta^2\bigr)
$$

よって全運動エネルギー

$$
T
= \frac{1}{2}(M+m)\dot x^2
  - m l\cos\theta\,\dot x\dot\theta
  + \frac{1}{2}m l^2\dot\theta^2
$$

---

### 4. 位置エネルギー $V$

重力加速度を $g$ とすると，重力による位置エネルギーは

$$
V = m g Y = m g l\cos\theta
$$

---

### 5. ラグランジュ関数

$$
L = T - V
  = \frac{1}{2}(M+m)\dot x^2
  - m l\cos\theta\,\dot x\dot\theta
  + \frac{1}{2}m l^2\dot\theta^2
  - m g l\cos\theta
$$

一般化座標を

$$
q_1 = x,\quad q_2 = \theta
$$

とする。ラグランジュ方程式は

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot q_i}
 - \frac{\partial L}{\partial q_i}
 = Q_i
$$

ここで

- $x$ 方向の一般化力：$Q_x = u$（カートに加える力）  
- $\theta$ 方向の一般化力：$Q_\theta = 0$（外トルクなし）  

---

### 6. カートの運動方程式（$x$ 方向）

$$
\frac{\partial L}{\partial x} = 0
$$

$$
\frac{\partial L}{\partial \dot x}
= (M+m)\dot x - m l\cos\theta\,\dot\theta
$$

時間微分すると

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot x}
= (M+m)\ddot x
  + m l\sin\theta\,\dot\theta^2
  - m l\cos\theta\,\ddot\theta
$$

ラグランジュ方程式より

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot x}
- \frac{\partial L}{\partial x}
= Q_x = u
$$

したがって

$$
(M+m)\ddot x
- m l\cos\theta\,\ddot\theta
+ m l\sin\theta\,\dot\theta^2
= u
$$

---

### 7. 振子の運動方程式（$\theta$ 方向）

$$
\frac{\partial L}{\partial \theta}
= m l\sin\theta\,\dot x\dot\theta
  + m g l\sin\theta
$$

$$
\frac{\partial L}{\partial \dot\theta}
= - m l\cos\theta\,\dot x
  + m l^2\dot\theta
$$

時間微分すると

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot\theta}
= m l\sin\theta\,\dot x\dot\theta
  - m l\cos\theta\,\ddot x
  + m l^2\ddot\theta
$$

ラグランジュ方程式より

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot\theta}
- \frac{\partial L}{\partial \theta}
= Q_\theta = 0
$$

すなわち

$$
m l\sin\theta\,\dot x\dot\theta
- m l\cos\theta\,\ddot x
+ m l^2\ddot\theta
- \bigl(
m l\sin\theta\,\dot x\dot\theta
+ m g l\sin\theta
\bigr)
= 0
$$

$\sin\theta\,\dot x\dot\theta$ の項が打ち消し合って

$$
m l^2\ddot\theta
- m l\cos\theta\,\ddot x
- m g l\sin\theta
= 0
$$

---

### 8. 最終的な運動方程式

したがって，倒立振子台車の非線形運動方程式は

$$
(M+m)\ddot x
- m l\cos\theta\,\ddot\theta
+ m l\sin\theta\,\dot\theta^2
= u,
$$

$$
m l^2\ddot\theta
- m l\cos\theta\,\ddot x
- m g l\sin\theta
= 0
$$


## 制御対象のモデル化：倒立振子を数式で表現する

u(t): システムに加える入力ベクトル。この場合は、カートに加える力。

A: システム行列。入力がないときのシステムの内部的な動きを決定する。

B: 入力行列。入力がシステムの状態をどのように変化させるかを決定する。

```python
import numpy as np
from scipy.linalg import solve_continuous_are
from control.matlab import lqr, care

# --- 第1章で解説するコード ---
# パラメータ
M = 1.0  # カートの質量
m = 0.1  # 振子の質量
l = 0.5  # 振子の長さ
g = 9.8  # 重力加速度

# A, B行列（状態空間モデル）
A = np.array([
    [0, 1, 0, 0],
    [0, 0, g*m/M, 0],
    [0, 0, 0, 1],
    [0, 0, g*(M+m)/(M*l), 0]
])
B = np.array([
    [0],
    [1/M],
    [0],
    [1/(M*l)]
])

# 行列AとBを表示
print("--- 行列A ---")
print(A)
print("\n--- 行列B ---")
print(B)

# 状態ベクトルx = [カート位置, カート速度, 振子角度, 振子角速度]^T
```

## 最適制御器の設計：LQRで倒立振子を立たせる

LQR（最適レギュレータ）は、状態空間モデル $\dot{x} = Ax + Bu$ に対して、状態フィードバック制御 $u = Fx$ を行う際の最適なフィードバックゲイン $F$ を決定する手法

<img src="png/16.png" alt="1" width="550" height="400">

図2　LQR制御ブロック線図

システムの出力（状態 $x$）をセンサで計測し、コントローラ（ゲイン $F$）を通して入力側に戻すことで、グルっと一周する「閉ループ」を形成する。

### $$J = \int_{0}^{\infty} \boldsymbol{x}(t)^{\top} \boldsymbol{Q} \boldsymbol{x}(t) + u(t)^{\top} \boldsymbol{R} u(t) \, \mathrm{d}t$$

この式は、「状態 $x$ をどれだけ速く0に収束させたいか（$Q$ で重み付け）」と、「制御入力 $u$ のエネルギーをどれだけ小さく抑えたいか（$R$ で重み付け）」のバランスを取るためのもの

#### 重み行列 $Q$ と $R$ の設定:

Q は、状態 $x$ に対する重み。np.diag([10, 1, 10, 1]) は、$x_1$ と $x_3$ の値を速く0に収束させることを、$x_2$ と $x_4$ よりも重視する設定を意味する。

R は、制御入力 $u$ に対する重み。この値を大きくすると、入力エネルギーを抑える（＝制御入力を小さくする）ことが優先される。

#### 教科書の理論では、最適ゲイン $F$ を求めるために、2つのステップを踏んでいる。P.182参照

----------------------------------------------------------------------------------------------------------------------------------------------------

#### ステップ1：リッカチ方程式を解く教科書の式(5.48)は「リッカチ方程式」と呼ばれるもので、この方程式を満たす特別な行列 $P$ を見つけることが最初のステップ。

##### $A^T P + PA - PBR^{-1}B^T P + Q = 0$ 　 ・・・　「　X = solve_continuous_are(A, B, Q, R)　」

solve_continuous_are という関数は、まさにこの「連続時間代数リッカチ方程式」を数値的に解くための専門の関数。

この関数に $A, B, Q, R$ を渡すことで、式(5.48)を満たす解 $P$ を計算する。つまり、コードの X が、教科書の $P$ に相当する。

----------------------------------------------------------------------------------------------------------------------------------------------------

#### ステップ2：最適ゲインを計算するステップ1で求めた行列 $P$ を使って、教科書の式(5.47)に基づき、最終的な最適ゲイン $F$ （教科書では $F_{opt}$）を計算する。

##### $F_{opt} = -R^{-1}B^T P$　　・・・　　「　F = -np.linalg.inv(R) @ B.T @ X　」

この1行は、式(5.47)の $F_{opt} = -R^{-1}B^T P$ を構成要素ごとに分解して計算している。

np.linalg.inv(R) は、$R$ の逆行列、すなわち $R^{-1}$ を計算している。

B.T は、$B$ の転置行列、すなわち $B^T$ を計算している。

X は、ステップ1で求めた $P$ 。

@ 記号は、行列の積（掛け算）を意味する。

これらを組み合わせると、コードは - (R^{-1}) @ (B^T) @ (X) となり、教科書の式(5.47) $F_{opt} = -R^{-1}B^T P$ と完全に一致する。

#### 「eigvals = np.linalg.eigvals(A + B @ F)」

閉ループ極の計算:　$u=Fx$ を元の式 $\dot{x} = Ax + Bu$ に代入すると、制御器を適用した後のシステム（閉ループ系）は $\dot{x} = (A + BF)x$ となる。このコードは、閉ループ系の行列 A + B @ F の固有値（極）を計算している。この固有値の実部がすべて負であれば、システムは安定となる。LQRで設計されたゲイン $F$ を使えば、閉ループ系は必ず安定になる。

```python
# --- 第2章で解説するコード ---
# Q, R（重み行列）
Q = np.diag([1, 10, 1, 10])
R = np.array([[1]])

# LQRで最適ゲインFを計算
X = solve_continuous_are(A, B, Q, R)
F = -np.linalg.inv(R) @ B.T @ X
eigvals = np.linalg.eigvals(A + B @ F)

# シミュレーション時間: 0秒から10秒までを1000分割
t = np.linspace(0, 10, 1000)


print("--- フィードバックゲイン F ---")
print(F)
print("--- 閉ループ極 ---")
print(eigvals)
```

## 近似線形モデルのデータ可視化

計算上で求めたフィードバックゲイン(F)や閉ループ極の安定性だけでは、実際のシステムの振る舞いを直感的に理解するのは困難なので、シミュレーションと結果の可視化をしていく。　

以下のプログラムを実行してください。

LQRによる状態フィードバック制御したときの、波形が出力されます。４つの状態（台車位置、台車速度、振子の角度、角速度）の応答グラフが示されます。

シミュレーションは以下の前提で行われています。

【シミュレーションの前提】
1. ４つの状態はすべて、誤差やノイズなく完全な状態で観測できている
2. 線形モデルで状態フィードバックを行っている。フィードバックのゲインFは先ほど設計した値である。

【確認事項】
1. グラフから、すべての状態が安定して０に収束していることを確認してください。
2. 状態はすべて完全にセンサ等で観測できるとしていますので、波形がきれいであることを確認してください。

```python
from scipy.signal import StateSpace, lsim
# グラフ描画ライブラリをインポートし、全体のデザインスタイルを設定する
import matplotlib.pyplot as plt


# 1. 閉ループシステムを作成
A_cl = A + B @ F
B_cl = np.zeros((4, 1))  # 外部からの入力はないためゼロ
C_cl = np.eye(4)         # 全ての状態を観測
D_cl = np.zeros((4, 1))
sys_cl = StateSpace(A_cl, B_cl, C_cl, D_cl)

# 2. シミュレーション条件を設定
# 初期条件: 振り子を5度(radに変換)傾けた状態
x0 = np.array([0, 0, np.deg2rad(5), 0])

# シミュレーション時間: 0秒から10秒までを1000分割 
#t = np.linspace(0, 10, 1000)

# 3. シミュレーションを実行
_, yout, _ = lsim(sys_cl, U=0*t, T=t, X0=x0)

# --- ▼▼ グラフ描画 ▼▼ ---
# 4. 描画エリアの作成
# 見た目を整えるためのスタイルシートを適用
plt.style.use('seaborn-v0_8-whitegrid')
# 2行2列のグラフエリアを作成。figが全体、axesが各プロットを指す
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
# グラフにメインタイトルを付ける
fig.suptitle('LQR', fontsize=16)


# 5. 各グラフの描画 (ループ処理)
# プロットデータとラベル
plot_data = [yout[:, 0], yout[:, 1], np.rad2deg(yout[:, 2]), yout[:, 3]]
titles = ['Cart Position (x)', 'Cart Velocity (x_dot)', 'Pendulum Angle (theta)', 'Pendulum Angular Velocity (theta_dot)']
ylabels = ['Position x [m]', 'Velocity dx/dt [m/s]', 'Angle theta [deg]', 'Angular Vel dθ/dt [rad/s]']

# 各サブプロットを描画
# zip: 複数のリストを同時にループ処理するための便利な関数
# zip関数でリストを束ね、forループで一度に処理
for ax, data, title, ylabel in zip(axes.ravel(), plot_data, titles, ylabels): 
    
    ax.plot(t, data) # x軸に時間t, y軸にシミュレーション結果dataをプロット
    ax.set_title(title) # サブプロットのタイトル設定
    ax.set_xlabel('Time [s]') # X軸ラベル設定
    ax.set_ylabel(ylabel) # Y軸ラベル設定
    ax.grid(True) # グリッド表示をオン

# 6. 仕上げと表示
plt.tight_layout()
fig.subplots_adjust(top=0.92)
plt.show()
```

## グラフ表示方法

コード内にも説明があります。

Pythonの matplotlib ライブラリを使い、シミュレーション結果をグラフに描画する。

### 1. 閉ループシステムを作成

・　A_cl = A + B @ F: 元のシステム A に、フィードバック制御 B @ F を足し合わせる。これにより、コントローラーの働き（状態xに応じて操作量uを決定するルール）がシステムの一部として組み込まれ、新しいシステム行列A_clが完成する。このA_clが、自律的に安定しようと振る舞うシステムの特性を表す。

・　np.zeros(...): シミュレーション中、外部から新たな力（目標値の変更や外乱など）は加えないため、入力uに関連するB行列、D行列はゼロになる。

・　C_cl = np.eye(4): C行列はどの状態を観測（出力）するかを決める。np.eye(4)（4x4の単位行列）を指定することで、「4つの状態量（位置, 速度, 角度, 角速度）を全て観測対象とする」という意味になる。

・　sys_cl = StateSpace(...): 4つの行列をscipyのStateSpaceクラスに渡すことで、シミュレーション可能なオブジェクトsys_clを作成する。

scipy.signal.StateSpaceは、システムの動的な振る舞いを定義する4つの行列（A, B, C, D）を、ひとまとめにして扱うためのツール

### 2. シミュレーション条件を設定

・　x0 = np.array([0, 0, np.deg2rad(5), 0]): これはシステムの初期状態x(0)のこと。[位置, 速度, 角度, 角速度]の順番に対応しており、「台車の位置と速度は0だが、振り子が最初に5度だけ傾いている」という状況を設定している。ここから制御がスタートする。

・　t = np.linspace(0, 10, 1000): これはシミュレーション時間の定義。「0秒から10秒後までを1000個の点で区切って観測する」という意味で、シミュレーションの長さと細かさを決めている。

### 3. シミュレーションを実行

・　lsim(sys_cl, U=0*t, T=t, X0=x0): scipyの線形シミュレーション関数lsimを呼び出している。

　・　sys_cl: 1で作成した制御対象のシステム。

　・　X0=x0: 2で設定した初期状態からスタート。

　・　T=t: 2で設定した時間に沿って計算。

　・　U=0*t: シミュレーション中の外部入力。今回は初期状態からの応答を見るだけのため、入力は常にゼロ。

## ローパスフィルタによるノイズ除去

### 1. なぜローパスフィルタ(LPF)が必要か？ (目的)

問題: 現実のセンサ（今回はカート位置  を測るセンサや、振子角度  を測るエンコーダ）は完璧ではなく、観測値には必ずノイズ（ランダムな誤差）が乗る。

もしノイズをそのまま使ったら？:

LQRコントローラは、このノイズを「本当に振子が細かく振動している」と勘違いしてしまう。u = F @ xこの x にノイズが含まれていると、制御入力 u（カートを押す力）もノイズに反応して激しく振動する。その結果、x（カート位置）が小刻みに振動する。また、アクチュエータ（モーター）に過剰な負荷がかかり、エネルギーを無駄遣いしたり、寿命を縮めたりする。

<img src="png/17.png" alt="1" width="650" height="500">

$$【再掲】図3　観測ノイズを含むフィードバック制御ブロック線図$$

観測ノイズを含むフィードバック制御系上図のように、観測した状態 $\boldsymbol{x}$ にノイズが乗ったままフィードバックゲイン $\boldsymbol{F}$ を掛けると、そのノイズが増幅されて制御入力 $\boldsymbol{u}$ に伝わり、システム全体を不安定にしてしまう。

目的:そこで、制御に使う前に、観測値から不要なノイズを除去する必要がある。このために、まずローパスフィルタ（LPF）を使う。

### 2. ローパスフィルタ (LPF) とは？

ローパスフィルタ（Low-Pass Filter, LPF）は、

 **低い周波数成分は通し，高い周波数成分を弱めるフィルタ**


#### (1) 周波数のイメージ

時間変化する信号には，

- ゆっくり変わる成分 … 低周波（例：台車や振子そのものの動き）
- チリチリ震える成分 … 高周波（例：センサノイズ）

が重なって含まれている。

LPFは，このうち**低周波成分をできるだけ残し，高周波成分を小さくする**働きをする。

#### (2) カットオフ周波数 \( f_c \) とは？

LPFには「どこまでを低周波として通すか」を決める  
**カットオフ周波数 \( f_c \) [Hz]** が設定されている。

- \( f_c \) より十分低い周波数成分  
  → ほぼそのまま通過（ゲイン ≒ 1）
- \( f_c \) を越える高い周波数成分  
  → 周波数が高いほど強く減衰

1次ローパスでは，カットオフ周波数 \( f_c \) において

- 振幅は直流の約 **0.707 倍（= 1/√2）**

となる。

#### (3) 制御での使いどころ

倒立振子や台車の制御では，

- 低周波成分：**実際の動き**
- 高周波成分：**センサノイズ**

とみなし，LPFによってノイズを抑えた信号を制御器に入力する，という目的で使う。

#### 次のシミュレーションでは、測定した実験データに対してオフラインでフィルタをかける。フィードバック制御に適用する前に「このフィルタを使っていたら、センサ値のノイズはどのくらい除去されるか」を安全に確認する。

```python
import numpy as np
from scipy.linalg import solve_continuous_are
from scipy.signal import StateSpace, lsim
import matplotlib.pyplot as plt



# --- 観測ノイズの追加（※ yout は変更しない ノイズはコメントなら前回のまま） ---
#noise_x_std_dev = 0.0      # x のノイズの標準偏差 [m]
#noise_theta_std_dev = 0.0  # theta のノイズの標準偏差 [rad]



# --- ローパスフィルタの適用 (lsimを使用) ---

# 1. フィルタの設計 (1次遅れ系: G(s) = wc / (s + wc))

# 2. フィルタシステムを作成する関数
def create_lpf_sys(fc):
    wc = 2 * np.pi * fc
    A = [[-wc]]
    B = [[wc]]
    C = [[1.0]]
    D = [[0.0]]
    return StateSpace(A, B, C, D)


# ノイズ生成
x_noise = np.random.normal(0, noise_x_std_dev, yout[:, 0].shape)
theta_noise = np.random.normal(0, noise_theta_std_dev, yout[:, 2].shape)

# ノイズを加えた「観測値」を別変数に保存
x_meas = yout[:, 0] + x_noise          # 観測された台車位置
theta_meas = yout[:, 2] + theta_noise  # 観測された振子角度

# --- 数値微分のための準備 ---
dt = t[1] - t[0]

# 台車速度と角速度の推定値（ローパス後の値からオイラー差分）
x_dot_euler_filtered = np.zeros_like(x_meas)
theta_dot_euler_filtered = np.zeros_like(theta_meas)

# それぞれのシステムを作成
sys_x = create_lpf_sys(fc_x)
sys_theta = create_lpf_sys(fc_theta)

# 3. lsimで個別にフィルタリング実行
# 位置 x (sys_x を使用)
_, x_filtered, _ = lsim(sys_x, U=x_meas, T=t)

# 角度 theta (sys_theta を使用)
_, theta_filtered, _ = lsim(sys_theta, U=theta_meas, T=t)

# オイラー差分： v[k] ≒ (x[k] - x[k-1]) / dt
x_dot_euler_filtered[1:] = (x_filtered[1:] - x_filtered[:-1]) / dt
theta_dot_euler_filtered[1:] = (theta_filtered[1:] - theta_filtered[:-1]) / dt

# 0 番目の点の扱い（ここでは 1 番目と同じにしておく）
x_dot_euler_filtered[0] = x_dot_euler_filtered[1]
theta_dot_euler_filtered[0] = theta_dot_euler_filtered[1]

# --- グラフ描画 (比較用) ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('LQR Control with Observation Noise & Low-Pass Filter', fontsize=16)

# プロットデータなどを整理
# [生データ(ノイズ入), フィルタ済データ, タイトル, Yラベル] のリスト
plot_configs = [
    # 位置 x
    {
        "ax": axes[0, 0],
        "raw": x_meas,                          # ノイズ入り位置
        "filtered": x_filtered.flatten(),       # LPF出力
        "title": 'Cart Position (x)',
        "ylabel": 'Position [m]'
    },
    # 速度 x_dot （LPF後の位置からオイラー差分で求めた速度）
    {
        "ax": axes[0, 1],
        "raw": x_dot_euler_filtered,           # 1次元のままでOK
        "filtered": None,
        "title": 'Cart Velocity (x_dot) from LPF+Euler',
        "ylabel": 'Velocity [m/s]'
    },
    # 角度 thetaz
    
    {
        "ax": axes[1, 0],
        "raw": np.rad2deg(theta_meas),         # ノイズ入り角度 [deg]
        "filtered": np.rad2deg(theta_filtered.flatten()),  # LPF出力 [deg]
        "title": 'Pendulum Angle (theta)',
        "ylabel": 'Angle [deg]'
    },
    # 角速度 theta_dot （LPF後の角度からオイラー差分で求めた角速度）
    {
        "ax": axes[1, 1],
        "raw": theta_dot_euler_filtered,       # 1次元のままでOK
        "filtered": None,
        "title": 'Pendulum Angular Velocity (theta_dot) from LPF+Euler',
        "ylabel": 'Angular Velocity [rad/s]'
    }
]

# ループ処理で各グラフを描画
for config in plot_configs:
    ax = config["ax"]
    
    # 生データ（ノイズあり or 推定値）
    ax.plot(t, config["raw"], color='lightgray', label='Raw / Noisy / Estimated', alpha=0.8)
    
    # フィルタ済みデータがある場合
    if config["filtered"] is not None:
        ax.plot(t, config["filtered"], linewidth=1.5, label='LPF Output')
    else:
        # 速度などはそのままのデータをメインの色で上書き
        ax.lines[0].set_color('tab:blue')
        ax.lines[0].set_alpha(1.0)

    ax.set_title(config["title"])
    ax.set_xlabel('Time [s]')
    ax.set_ylabel(config["ylabel"])
    ax.legend()
    ax.grid(True)

plt.tight_layout()
fig.subplots_adjust(top=0.92)
plt.show()
```

### ※上記プログラムの説明は以下の「・・・」をクリック


### 1 伝達関数による定義
今回使用する「1次遅れフィルタ」は、入力 $U(s)$ と出力 $Y(s)$ の関係を以下の伝達関数 $G(s)$ で表す。

$$
G(s) = \frac{Y(s)}{U(s)} = \frac{1}{Ts + 1} = \frac{\omega_c}{s + \omega_c}
$$

* $T$: **時定数** [s]
* $\omega_c = 1/T $: **カットオフ角周波数** [rad/s]
* $f_c = \frac{\omega_c}{2\pi} = \frac{1}{2\pi T} $: **カットオフ周波数** [Hz]
* このシステムは、低周波の成分ほど、そのまま通過（ローパス）し、高周波となるほど減少させる。カットオフ周波数$f_c$ では約30%程信号の振幅が減衰する。



### 2 状態空間モデルへの変換
コンピュータでシミュレーション（`lsim`）を行うために、上記の式を**状態空間モデル** $\dot{x} = Ax + Bu, \quad y = Cx + Du$ に変換する。

伝達関数を微分方程式に戻すと $\dot{y} = -\omega_c y + \omega_c u$ となるため、フィルタの状態変数を $x_f$ とすると、係数行列は以下のようになる。

$$
A = [-\omega_c], \quad B = [\omega_c], \quad C = [1], \quad D = [0]
$$

コード内の `create_lpf_sys(fc)` 関数は、この行列計算を自動化している部分。

---

## 3. Pythonコードの解説：`lsim` とは？

今回のコードの核心部分

#### lsimで個別にフィルタリング実行
_, x_filtered, _ = lsim(sys_x, U=yout[:, 0], T=t)

関数 scipy.signal.lsim の役割
    
lsim (Linear simulation) は、線形システムに任意の入力を与えたときの応答を計算する関数。

入力:

sys: フィルタのシステムモデル（$A,B,C,D$ 行列）
    
U: ノイズを含んだセンサデータの配列（時系列データ）

T: 時間配列

処理:　内部で微分方程式 $\dot{x}_f = Ax_f + Bu$ を数値積分し、「もしこのフィルタ回路に信号を通したら、どういう波形が出てくるか」をシミュレートする。

出力:　戻り値に、ノイズが除去されたきれいなデータが入る。

## ローパスフィルタを制御系に挿入する

シミュレーションの構成は上図の通り。観測ノイズを含んだ信号がまずローパスフィルタを通過し、その滑らかになった信号をもとに数値微分で速度を推定する。これらを合わせた「推定状態ベクトル」に対してLQRゲイン $\boldsymbol{F}$ がかかる。

フィルタがループ内にあるため、フィルタによる位相遅れ（反応の遅れ）がシステムの安定性に影響する点に注目してください。

【確認事項】
1. カットオフ周波数を両方とも20Hzに設定すると、ローパスフィルタの影響が小さくなり、安定ではあるがノイズによる振動、特に数値微分で求めている速度の値が大きく振動することを確認してください。
2. ローパスフィルタのカットオフ周波数が fc_x =0.5 Hz fc_theta=0.5 Hzでは、不安定になることを確認してください。

```python
# ローパスフィルタのカットオフ周波数(Hz)を設定
fc_x = 0.5     # 位置x用 
fc_theta = 0.5  # 角度θ用 
```

```python
import numpy as np
import matplotlib.pyplot as plt


# カットオフ周波数 [Hz]
#fc_x = 0.5      # カート位置 x 用#
#fc_theta = 0.5   # 振子角度 theta 用

#######
# 時間ベクトルとサンプル時間　※これまでと同じなら変える必要なし
#######
t_common = t
dt_common = dt

# =========================================================
# 1. ローパスフィルタ係数（前進オイラーで1次遅れを近似）
#    ※ t_common, dt_common は共通セルで定義済みを想定
# =========================================================
def calc_alpha_fe(fc, dt):
    """カットオフ周波数 fc[Hz] から、オイラー差分用の係数 alpha を計算"""
    omega = 2 * np.pi * fc
    return omega * dt


alpha_x = calc_alpha_fe(fc_x, dt_common)
alpha_theta = calc_alpha_fe(fc_theta, dt_common)

if alpha_x >= 1.0 or alpha_theta >= 1.0:
    print("警告: オイラー差分が不安定になる可能性があります (α >= 1)")


# =========================================================
# 2. 観測ノイズ（位置 x と 角度 theta にのみ加える） 以前のままならコメントアウト
# =========================================================
# 観測ノイズ
# --- 観測ノイズの追加（※ yout は変更しない） ---
#noise_x_std_dev = 0.001      # x のノイズの標準偏差 [m]
#noise_theta_std_dev = 0.001  # theta のノイズの標準偏差 [rad]

common_noise_x = np.random.normal(0.0, noise_x_std_dev, len(t_common))
common_noise_theta = np.random.normal(0.0, noise_theta_std_dev, len(t_common))
    

# =========================================================
# 3. シミュレーションの準備
#    状態: x = [x, x_dot, theta, theta_dot]^T
# =========================================================
# 初期状態：x=0, x_dot=0, theta=5deg, theta_dot=0
x0 = np.array([0.0, 0.0, np.deg2rad(5.0), 0.0])
x = x0.copy()

history = {
    't': t_common,
    'x_true':     np.zeros((len(t_common), 4)),  # 真の状態
    'x_observed': np.zeros((len(t_common), 4)),  # ノイズ付き観測
    'x_filtered': np.zeros((len(t_common), 4)),  # フィルタ＋速度推定状態
}

# フィルタ内部状態（位置・角度）
x_filt_pos = x[0]      # カート位置のフィルタ内部状態
theta_filt = x[2]      # 角度のフィルタ内部状態
x_observed_prev = x0.copy()  # 前ステップの観測値

# =========================================================
# 4. シミュレーションループ
#    - x, theta にノイズ
#    - 観測 → LPF → フィルタ後の x, theta
#    - その差分から x_dot, theta_dot をオイラー差分で推定
# =========================================================
for i, t_c in enumerate(t_common):
    # --- 真の状態を記録 ---
    history['x_true'][i] = x

    # --- 観測（ノイズ付与） ---
    noise_x = common_noise_x[i]
    noise_theta = common_noise_theta[i]

    x_observed = x.copy()
    x_observed[0] += noise_x        # 位置 x にノイズ
    x_observed[2] += noise_theta    # 角度 theta にノイズ
    history['x_observed'][i] = x_observed

    # --- ローパスフィルタ（位置 x, 角度 theta）---
    # 一つ前の観測値を入力として1次遅れをオイラーで近似
    x_filt_pos_new = (1 - alpha_x) * x_filt_pos   + alpha_x * x_observed_prev[0]
    theta_filt_new = (1 - alpha_theta) * theta_filt + alpha_theta * x_observed_prev[2]

    # --- フィルタ後の位置・角度からオイラー差分で速度推定 ---
    x_dot_est     = (x_filt_pos_new - x_filt_pos) / dt_common       # カート速度
    theta_dot_est = (theta_filt_new - theta_filt) / dt_common       # 角速度

    # フィルタ内部状態を更新（次ステップ用）
    x_filt_pos = x_filt_pos_new
    theta_filt = theta_filt_new

    # 次ステップのフィルタ入力用に観測を保存
    x_observed_prev = x_observed.copy()

    # --- 制御用の推定状態ベクトルを構成 ---
    # 0: フィルタ後の位置 x
    # 1: フィルタ後の x からオイラー差分で推定した x_dot
    # 2: フィルタ後の角度 theta
    # 3: フィルタ後の theta からオイラー差分で推定した theta_dot
    x_for_control = np.zeros_like(x)
    x_for_control[0] = x_filt_pos
    x_for_control[1] = x_dot_est
    x_for_control[2] = theta_filt
    x_for_control[3] = theta_dot_est

    history['x_filtered'][i] = x_for_control

    # --- LQR制御入力の計算 ---
    # ※ F は共通セルで定義済み (u = F @ x_for_control)
    u = F @ x_for_control          # u はスカラー or (1,) を想定
    u = u.item()                   # 念のためスカラー化

    # --- 真の状態の時間発展（オイラー積分）---
    # ※ A, B も共通セルで定義済み
    dx = A @ x + B.flatten() * u   # B が (4,1) でも対応できるよう flatten
    x = x + dx * dt_common

# =========================================================
# 5. グラフ描画
# =========================================================
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('LQR Control with LPF & Velocity from Filtered Position/Angle', fontsize=16)

# --- カート位置 x ---
ax = axes[0, 0]
ax.plot(t_common, history['x_true'][:, 0],     label='True x',              color='green',  linewidth=2)
ax.plot(t_common, history['x_observed'][:, 0], label='Observed x (Noisy)',  color='orange', alpha=0.5)
ax.plot(t_common, history['x_filtered'][:, 0], label=f'Filtered x (fc={fc_x}Hz)', color='blue', linewidth=2)
ax.set_title('Cart Position x')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Position [m]')
ax.legend()
ax.grid(True)

# --- カート速度 x_dot（真値 vs 推定）---
ax = axes[0, 1]
ax.plot(t_common, history['x_true'][:, 1],     label='True x_dot',                         color='green', linewidth=2)
ax.plot(t_common, history['x_filtered'][:, 1], label='Estimated x_dot (from filtered x)',  color='blue',  linewidth=2)
ax.set_title('Cart Velocity x_dot')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Velocity [m/s]')
ax.legend()
ax.grid(True)

# --- 振子の角度 theta ---
ax = axes[1, 0]
ax.plot(t_common, np.rad2deg(history['x_true'][:, 2]),     label='True theta',              color='green',  linewidth=2)
ax.plot(t_common, np.rad2deg(history['x_observed'][:, 2]), label='Observed theta (Noisy)',  color='orange', alpha=0.5)
ax.plot(t_common, np.rad2deg(history['x_filtered'][:, 2]), label=f'Filtered theta (fc={fc_theta}Hz)', color='blue', linewidth=2)
ax.set_title('Pendulum Angle θ')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Angle [deg]')
ax.legend()
ax.grid(True)

# --- 振子の角速度 theta_dot（真値 vs 推定）---
ax = axes[1, 1]
ax.plot(t_common, history['x_true'][:, 3],     label='True theta_dot',                          color='green', linewidth=2)
ax.plot(t_common, history['x_filtered'][:, 3], label='Estimated theta_dot (from filtered θ)',   color='blue',  linewidth=2)
ax.set_title('Pendulum Angular Velocity θ_dot')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Angular Velocity [rad/s]')
ax.legend()
ax.grid(True)

plt.tight_layout()
fig.subplots_adjust(top=0.92)
plt.show()
```

### どうやって？ 

(一次LPFの例):今回のコードで使ったのは「一次ローパスフィルタ」と呼ばれる、最もシンプルで一般的なフィルタ。

$y[k] = \alpha \cdot x[k] + (1 - \alpha) \cdot y[k-1]$

$x[k]$: 現在の観測値（ノイズ付き）

$y[k-1]$: 1ステップ前のフィルタ結果（過去の値）

$y[k]$: 今回のフィルタ結果（新しい推定値）

$\alpha$: フィルタ係数（$0 \le \alpha \le 1$）

これは「現在の観測値と過去の推定値を、αで重み付け平均している」ことを意味する。αが小さいほど、過去の値を重視する「なまった（＝強い）フィルタ」になり、fcは低くなる。

----------------------------------------------------------------------------------------------------------------------------------------------------

### 1. 連続時間モデルローパスフィルタは、時定数 $T$ を持つ以下の一次遅れ系で表される。

### $L(s) = \frac{Y(s)}{U(s)} = \frac{1}{Ts + 1}$

$U(s)$: フィルタへの入力（ノイズを含んだ観測値）

$Y(s)$: フィルタからの出力（ノイズが除去された値）

$T$: フィルタの時定数。ご教示の通り、カットオフ周波数 

$f_c$ [Hz] とは $T = \frac{1}{2\pi f_c} = \frac{1}{\omega_c}$ の関係がある。この伝達関数を、時間領域の微分方程式に変換する。

$Y(s) (Ts + 1) = U(s)$

$T \cdot sY(s) + Y(s) = U(s)$

これを逆ラプラス変換すると（$\dot{y}(t) \leftarrow sY(s)$）、以下の微分方程式が得られる。

$T \dot{y}(t) + y(t) = u(t)$

#### $\dot{y}(t) = \frac{-y(t) + u(t)}{T}$ ... 1

----------------------------------------------------------------------------------------------------------------------------------------------------

### 2. 離散化（理論からコードへ）

次に、この連続的な微分方程式を、シミュレーションで使える離散的な式（漸化式）に変換する。

コードの関数名「 calc_alpha_fe 」にある通り、オイラー前進差分を用いる。

オイラー前進差分は、微分 $\dot{y}(t)$ を以下の差分で近似する。

#### $$\dot{y}(t_k) \approx \frac{y[k+1] - y[k]}{T_s}$$

$T_s$: サンプリング周期（コードの dt）

$y[k]$: 時刻 $t_k$ でのフィルタ出力値

$y[k+1]$: 次の時刻 

$t_{k+1}$ でのフィルタ出力値この近似式を、1の微分方程式に代入する。

#### $$\frac{y[k+1] - y[k]}{T_s} = \frac{-y[k] + u[k]}{T}$$

この式を、求めたい次のステップの値 $y[k+1]$ について解く。

$y[k+1] - y[k] = \frac{T_s}{T} (-y[k] + u[k])$

$y[k+1] = y[k] - \frac{T_s}{T} y[k] + \frac{T_s}{T} u[k]$

$y[k+1] = \left(1 - \frac{T_s}{T}\right) y[k] + \frac{T_s}{T} u[k]$

ここで、$\frac{T_s}{T}$ = α(alpha) と定義すると、

上記の式は、

#### $$y[k+1] = \left(1 - α\right) y[k] + α * u[k]$$

と表せる。

また、$T = \frac{1}{2\pi f_c}$ (fc:カットオフ周波数 Hz)と表せる。

#### (参考)

$\frac{T_s}{T}$ = α(alpha) : 現在の入力

$\left(1 - \frac{T_s}{T}\right)$ = 1 - α : 「現在の出力 y[k] を、次の出力 y[k+1] にどれだけ引き継ぐか」という割合。αが小さいほど、この値が一に近づき、過去の値を反映させる(フィルタが強い)ことを意味する。

## コード内解説

#### ① 係数 $\alpha$ の計算   arufaの説明

コードでは、alpha = omega * dt という計算をしている。

$$omega = 2\pi f_c = \omega_cdt = T_s$$

ここで、理論式の時定数 $T$ は $T = \frac{1}{\omega_c}$ だから、

$$\alpha = \omega_c \cdot T_s = \left(\frac{1}{T}\right) \cdot T_s = \frac{T_s}{T}$$

となり、コードの alpha は、理論式中の $\frac{T_s}{T}$ と全く同じものであることがわかる。

#### ② 漸化式の計算

理論式（ステップ2）に $\alpha = \frac{T_s}{T}$ を代入すると、

$$y[k+1] = (1 - \alpha) y[k] + \alpha u[k]$$

コードのシミュレーションループでは、以下の計算を行っている

#### y[k] = (1 - alpha) * y[k-1] + alpha * x[k-1] （コメントの式）
#### x_filtered = (1 - alpha_x) * x_filtered + alpha_x * x_observed_prev[2]

これは、$y[k] = (1 - \alpha)y[k-1] + \alpha u[k-1]$ を計算している。

「現在のフィルタ値 $y[k]$ は、1つ前のフィルタ値 $y[k-1]$ と1つ前の観測値 $u[k-1]$ から計算する」という、形になっている。

## カルマンフィルタ（定常カルマンフィルタ）

#### なぜカルマンフィルタ(KF)が必要か？

これまでのシミュレーションでは、以下のような問題が存在した。

1. LPF（ローパスフィルタ）の限界 : ローパスフィルタは「過去の平均」を取る仕組みであるため、ノイズを強く消そうとすると必ず **「位相遅れ（反応の遅れ）」** が発生し、倒立振子のような不安定なシステムでは致命的となる。

2. 数値微分の困難さ : センサ値に基づいて数値微分を行うとノイズの影響を取り除くことは極めて難しい。

カルマンフィルタは、「モデルを使って未来を予測」することで、位相遅れの発生を抑えつつノイズを除去できる。さらに、センサで直接測っていない状態（速度 $\dot{x}$ や角速度 $\dot{\theta}$）も、物理法則に基づいて推定できる。このため多くの場合、上記の問題を解決できる可能性がある。

#### カルマンフィルタとは何か？

(原理)KFは、システムの「動作モデル」を利用して、観測データから「真の状態」を最適に推定する**状態推定器（オブザーバ）**。

KFは、以下の2つの情報を利用する。

システムのモデル ($A, B, C$) : 「振子がどう動くか」「何を測定するか」の定義。

ノイズの統計量 ($Q_N, R_N$):

　・$R_N$ (観測ノイズ): センサが「どれくらい信用できないか」
 
　・$Q_N$ (プロセスノイズ): モデルが「どれくらい信用できないか」（モデル化していない外乱や誤差）
 
 システムの状態空間モデルの定数（A,B,C,D）が時不変の場合、上記のノイズの影響を最適に除去するカルマンフィルタは「定常カルマンフィルタ」と呼ばれる。

## 定常カルマンフィルタ

### 1. 定常ゲインの事前計算（ループの外）

このコードでは、シミュレーションループに入る前に、control.matlab.lqe 関数を使って「最適なゲイン $L$」を一度だけ計算している。これが「定常カルマンフィルタ」の特徴。

#### 数式 (リカッチ方程式の解):　コスト関数（誤差の分散）を最小化するゲイン $L$ を求める。

#### $$A P + P A^T - P C^T R^{-1} C P + Q = 0$$

#### $$L = P C^T R^{-1}$$

教科書 p.243

#### コード: **L, P, E = lqe(A, G_matrix, C, QN, RN)**

・A: システム行列

・QN, RN: プロセスノイズ分散 $Q$、観測ノイズ分散 $R$

・L: 計算された 定常カルマンゲイン

#### 意味:

・システムモデル ($A, G_{matrix}, C$) とノイズ統計量 ($Q_N, R_N$) から、リカッチ方程式を解いて 定常カルマンゲイン L を求める。

・この L が、ループ内での「予測と観測のブレンド比率」を決定する固定値になる。

---------------------------------------------------------------------------------------------------------------------------------------------------

### 2. 推定の実行（ループの中）

ループ内では、以下の 推定値の微分方程式 を計算している。この1行の式の中に、「予測」と「更新」の両方の要素が含まれている。ここで　$y$ は観測値であり, $y = Cx$ でモデル化されている。

#### 数式: $$\dot{\hat{x}} = ( \dot{\hat{x}}_{pred} + \dot{\hat{x}}_{corr} ) = \underbrace{A \hat{x} + B u}_{\text{(a) 予測}} + \underbrace{L (y - C \hat{x})}_{\text{(b) 更新}}$$ 

コードでは以下のように記述されている。

#### コード: dx_hat = A @ x_hat + B.flatten() * u + L @ (y - C @ x_hat)

---------------------------------------------------------------------------------------------------------------------------------------------------

### (a) 予測

「物理モデルに従うと、次の瞬間どう動くか」を計算する部分。

#### 数式:　$$\dot{\hat{x}}_{pred} = A \hat{x} + B u$$ 

コードでは以下のように記述されている。

#### コード: A @ x_hat + B.flatten() * u

役割: モデル（運動方程式）に基づいて状態の変化速度を予測する。これにより、センサの遅れに依存せず、位相遅れのない推定が可能になる。

---------------------------------------------------------------------------------------------------------------------------------------------------

### (b) 更新 

「実際の観測値と予測値のズレ（残差）を使って、軌道を修正する」部分。

#### 数式:　$$\dot{\hat{x}}_{corr} = L (y - C \hat{x})$$

コードでは以下のように記述されている。

#### コード: L @ (y - C @ x_hat)

・y: 実際の観測値（ノイズ入り） $\to y$

・C @ x_hat: 現在の推定値から予想される観測値 

・(y - C @ x_hat): 観測残差（イノベーション）。予測と現実のズレ。

役割:　この「ズレ」に信頼度（ゲイン L）を掛けて、予測の軌道を現実に引き戻すように修正する。

```python
import numpy as np
from scipy.linalg import solve_continuous_are
from control.matlab import lqe 
import matplotlib.pyplot as plt

#######
# 時間ベクトルとサンプル時間　※これまでと同じなら変える必要なし
#######
#t_common = t
#dt_common = dt

# ===== 1. システム定義 (倒立振子) =====
M = 1.0; m = 0.5; l = 1.0; g = 9.8
# 状態空間モデル dx/dt = A x + B u
A = np.array([
    [0, 1, 0, 0],
    [0, 0, -m*g/M, 0],
    [0, 0, 0, 1],
    [0, 0, (M+m)*g/(M*l), 0]
])
B = np.array([[0], [1/M], [0], [-1/(M*l)]])
C = np.array([[1, 0, 0, 0], [0, 0, 1, 0]]) # xとθを観測 (2次元の出力)

# ===== 2. LQR設計 =====
Q_lqr = np.diag([10, 1, 10, 1])
R_lqr = np.array([[1]])
X_lqr = solve_continuous_are(A, B, Q_lqr, R_lqr)
F_lqr = -np.linalg.inv(R_lqr) @ B.T @ X_lqr
# F_lqr は u = -F x のゲイン (1x4行列)

# ===== 3. カルマンフィルタ設計 =====
#noise_x_std_dev = 0.05      # 位置の観測ノイズ標準偏差
#noise_theta_std_dev = 0.05  # 角度の観測ノイズ標準偏差


#RN = np.diag([noise_x_std_dev**2, noise_theta_std_dev**2]) # 観測ノイズ共分散行列 R
RN = np.diag([0.0001, 0.0001])                       # 観測ノイズ共分散行列 R
QN = np.diag([0.01, 0.01, 0.001, 0.001])             # プロセスノイズ共分散行列 Q
G_matrix = np.eye(4)

# リッカチ方程式を解いてカルマンゲイン L を求める
L, P, E = lqe(A, G_matrix, C, QN, RN)

# ===== 4. シミュレーションの準備 =====
#dt = 0.005
#t_common = np.arange(0, 10, dt)

# 初期状態
x_true = np.array([0.0, 0.0, 0.1, 0.0]) # 真の状態 (0.1rad = 約5.7度)
x_hat  = np.array([0.0, 0.0, 0.0, 0.0]) # 推定値の初期値

# 結果を保存するための辞書 (グラフ描画コードに合わせる)
history = {
    't': t_common,
    'x_true': np.zeros((len(t_common), 4)),      # 真の状態
    'y_observed': np.zeros((len(t_common), 2)),  # 観測値 (x, thetaのみ)
    'x_hat': np.zeros((len(t_common), 4)),       # カルマンフィルタ推定値
}

# ===== 5. シミュレーションループ =====
for ii in range(len(t_common)):
    # --- 履歴の記録 ---
    history['x_true'][ii] = x_true
    history['x_hat'][ii]  = x_hat
    
    # --- 観測の生成 (ノイズ付加) ---
    noise = np.array([
        np.random.normal(0, noise_x_std_dev),
        np.random.normal(0, noise_theta_std_dev)
    ])
    # y = Cx + v
    y = C @ x_true + noise
    history['y_observed'][ii] = y

    # --- 制御入力の計算 (推定値 x_hat を使用) ---
    # u = -F * x_hat (F_lqrは正のフィードバックゲインとして計算されている場合符号に注意、ここでは一般的 u = (A-BK)x の形式として F_lqr @ x_hat とする)
    # 元のコードに合わせて (F_lqr @ x_hat).item() とします
    u = (F_lqr @ x_hat).item()
    
    # --- カルマンフィルタの更新 (推定) ---
    # d(x_hat)/dt = A*x_hat + B*u + L*(y - C*x_hat)
    # 予測 + 更新
    y_tilde = y - C @ x_hat # イノベーション (観測残差)
    dx_hat = A @ x_hat + B.flatten() * u + L @ y_tilde
    x_hat += dx_hat * dt

    # --- 真の状態の更新 (物理) ---
    dx_true = A @ x_true + B.flatten() * u
    x_true += dx_true * dt


# ===== 6. グラフ描画 (指定のスタイルに統合) =====
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Kalman Filter (Noise on x and θ)', fontsize=16)

# --- (1) カート位置 x ---
ax = axes[0, 0]
ax.plot(t_common, history['x_true'][:, 0], label='True Value', color='green', linewidth=2)
ax.plot(t_common, history['y_observed'][:, 0], label='Observed (Noisy)', color='orange', alpha=0.5)
ax.plot(t_common, history['x_hat'][:, 0], label='KF Estimate', color='blue', linewidth=2)
ax.set_title('Cart Position (x)')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Position [m]')
ax.legend()
ax.grid(True)

# --- (2) カート速度 x_dot ---
ax = axes[0, 1]
ax.plot(t_common, history['x_true'][:, 1], label='True Value', color='green', linewidth=2)
# 速度は直接観測していないため、観測プロットはなし
ax.plot(t_common, history['x_hat'][:, 1], label='KF Estimate', color='blue', linewidth=2)
ax.set_title('Cart Velocity (x_dot)')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Velocity [m/s]')
ax.legend()
ax.grid(True)

# --- (3) 振子角度 θ ---
ax = axes[1, 0]
# ラジアン -> 度 に変換してプロット
ax.plot(t_common, np.rad2deg(history['x_true'][:, 2]), label='True Value', color='green', linewidth=2)
ax.plot(t_common, np.rad2deg(history['y_observed'][:, 1]), label='Observed (Noisy)', color='orange', alpha=0.5)
ax.plot(t_common, np.rad2deg(history['x_hat'][:, 2]), label='KF Estimate', color='blue', linewidth=2)
ax.set_title('Pendulum Angle (theta)')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Angle [deg]')
ax.legend()
ax.grid(True)

# --- (4) 振子角速度 θ_dot ---
ax = axes[1, 1]
ax.plot(t_common, history['x_true'][:, 3], label='True Value', color='green', linewidth=2)
# 角速度は直接観測していないため、観測プロットはなし
ax.plot(t_common, history['x_hat'][:, 3], label='KF Estimate', color='blue', linewidth=2)
ax.set_title('Pendulum Angular Velocity (theta_dot)')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Angular Velocity [rad/s]')
ax.legend()
ax.grid(True)

plt.tight_layout()
fig.subplots_adjust(top=0.94)
plt.show()
```
